#### _1.3_
Junta las columnas "Fills" y "No Fills" en una sola columna, utilizando el método .apply() y definiendo una función que resuelva el problema. La columna nueva debe llamarse "Hijos" y tomar los valores "Sí" o "No".

**NOTA**: Para comprobar coherencia que cuando Fills tiene valor, columna No fills es NaN, y sacar una lista de los registros contradictorías:  
df_sprint10[(df_sprint10["Fills"] == 1) & (df_sprint10["No Fills"].notna())]

```python
# COMPROBACIONES:
print(df_sprint10[['Fills', 'No Fills']].info())
df_comprobacion = df_sprint10.loc[df_sprint10["Fills"] == 1, "No Fills"].isna().all()
df_comprobacion
```

```python
def clasificar_fills_si_no(valores_fila):
    """
    Combina las columnas 'Fills' y 'No Fills' en una sola y actualiza valor según:
    Si valor en 'Fills': "Sí"
    Si valor en 'No Fills': "No"
    Return: str con si o no
    """
    if pd.notnull(valores_fila["Fills"]):
        return "Sí"
    else:
        return "No"

df_sprint10["Fills"] = df_sprint10.apply(clasificar_fills_si_no, axis=1)
df_sprint10 = df_sprint10.drop(columns=["No Fills"])
print(df_sprint10["Fills"].value_counts())
df_sprint10.sample(10)

```

#### _1.4_
Crea una taula resum que permeti veure el sou mig, medià, mínim i màxim per Gènere.
Ordena la taula en funció del sou mig.

```python
df_sprint10['Salari mensual'] = df_sprint10['Salari mensual'].str.replace(' €', '').str.replace('.', '').astype(int)

df_resum_salario = (
    df_sprint10
        .groupby("Gènere")["Salari mensual"]
        .agg(
            salario_promedio = "mean",
            mediana_salario  = "median",
            salario_mínimo   = "min",
            salario_máximo   = "max"
        )
)
df_resum_salario = df_resum_salario.round(2)
print(df_resum_salario)
```

#### _1.6_
Crea una columna nova que sigui la data de naixament en format Datetime a partir de les columnes dia, mes i any.  
Utilitzant aquesta columna crea una funció que donada una data, et calculi l'edat actual a dia d'avui.  
Utilitza la funció que acabes de crear per generar una columna nova al DataFrame amb l'edat actual.

```python
from datetime import datetime

df_sprint10['Data Naixement'] = pd.to_datetime(
    df_sprint10[['Any de Naixement', 'Mes de Naixement', 'Dia de Naixement']].rename(
        columns={'Any de Naixement': 'year', 
                 'Mes de Naixement': 'month', 
                 'Dia de Naixement': 'day'}
    )
)

def calcular_edad(fecha_nacimiento):
    hoy = datetime.now()
    edad = hoy.year - fecha_nacimiento.year - (
        (hoy.month, hoy.day) < (fecha_nacimiento.month, fecha_nacimiento.day)
    )
    return edad

df_sprint10['Edat Actual'] = df_sprint10['Data Naixement'].apply(calcular_edad)

print(df_sprint10[['Nom', 'Data Naixement', 'Edat Actual']])
```

```python
df_increment = pd.DataFrame({
    "Grup":["Grup A","Grup B","Grup C", "Grup D" ],
     "Increment":["5%","3,5%","2%","8%"]
     })

df_increment
```

```python
df_sprint10 = pd.merge(
    df_sprint10,
    df_increment,
    left_on="Grup Professional",
    right_on="Grup",
    how="left"
)
```

```python
df_sprint10["Increment_float"] = (
    df_sprint10["Increment"]
        .astype(str)                       
        .str.replace("%", "", regex=False)   
        .str.replace(",", ".", regex=False)  
        .astype(float)                       
        / 100                                
)

df_sprint10["Salari mensual"] = (
    df_sprint10["Salari mensual"]
        .astype(str)                       
        .str.replace(".", "", regex=False)   
        .str.replace(",", ".", regex=False)  
        .str.replace("€", "", regex=False)   
        .str.strip()                         
        .astype(float)
)

df_sprint10["Salari mensual"].dtype
df_sprint10["Increment_float"].dtype
```

```python
df_sprint10["Salari mensual actualizat"] = (df_sprint10["Salari mensual"] * (1 + df_sprint10["Increment_float"])).round(2)

#Se podría borrar la columna "INCREMENT" con el % o dejarla como apoyo visual para informes o visualizaciones.
```

#### 2.2
Utilizando un bucle, exporta en 4 archivos (formato .xlsx o .csv) los datos de cada Grupo Profesional. Por ejemplo: "datos_GrupA.xlsx" , "datos_GrupB.xlsx" ...

```python
grupos_profesionales = df_sprint10["Grup Professional"].unique()

for grup in grupos_profesionales:
    df_grupo = df_sprint10[df_sprint10["Grup Professional"] == grup]
    nom_doc = f"datos_{grup}.xlsx"
    df_grupo.to_excel(nom_doc, index=False)
    print(f"Exportado: {nom_doc}")
```

Exporta un 5º DataFrame en formato .xlsx o .csv que contenga cuántos trabajadores hay por cada Grupo Profesional, cuál es su sueldo medio y cuál es su edad media.

```python
resum_grups = df_sprint10.groupby("Grup Professional").agg(
    Num_treballadors=("Nom", "count"),
    Salari_mig=("Salari mensual", "mean"),
    Edat_mitjana=("Edat Actual", "mean")
).reset_index()


resum_grups["Salari_mig"] = resum_grups["Salari_mig"].round(2)
resum_grups["Edat_mitjana"] = resum_grups["Edat_mitjana"].round(0)

resum_grups.columns = ["Grup Professional", "Num de treballadors", "Salari mitjà Eur", "Edat mitjana"]

resum_grups.to_excel("resum_grups_professionals.xlsx", index=False)
resum_grups
```
